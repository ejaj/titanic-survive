# preprocessing pipelines

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
```

***pipeline for numerical attributes***

```python
num_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])
```

***pipeline for the categorical attributes***

```python
cat_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("cat_encoder", OneHotEncoder(sparse=False)),
])
```

***join the numerical and categorical pipelines***

```python
from sklearn.compose import ColumnTransformer
```

```python
num_attribs = ["Age", "SibSp", "Parch", "Fare"]
cat_attribs = ["Pclass", "Sex", "Embarked"]
```

```python
preprocess_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("cat", cat_pipeline, cat_attribs),
])
```

```python
X_train = preprocess_pipeline.fit_transform(df_train[num_attribs + cat_attribs])
X_train
```

```python
y_train = df_train["Survived"]
```

# Train the model

***Random Forest***

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
forest_clf = RandomForestClassifier(n_estimators=100, random_state=42)
forest_clf.fit(X_train, y_train)
```

```python
X_test = preprocess_pipeline.transform(df_test[num_attribs + cat_attribs])
y_pred = forest_clf.predict(X_test)
```

***Now we use cross-validation to have an idea of how good our model***

```python
from sklearn.model_selection import cross_val_score
```

```python
forest_scores = cross_val_score(forest_clf, X_train, y_train, cv=10)
forest_scores.mean()
```

***SVC***

```python
from sklearn.svm import SVC
```

```python
svm_clf = SVC(gamma="auto")
svm_clf
```

```python
svm_scores = cross_val_score(svm_clf, X_train, y_train, cv=10)
svm_scores.mean()
```
